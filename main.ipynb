# Nvidia Stock Price Analysis (1999–Present)

## Introduction
This project analyzes the historical stock price performance of **Nvidia Corporation (NVDA)**. As a key player in the semiconductor and AI industries, Nvidia's stock trajectory offers valuable insights into market trends and technological shifts over the past two decades.

## Objective
The goal of this notebook is to:
1.  **Extract** historical stock data using the Yahoo Finance API (`yfinance`).
2.  **Process** and clean the data for analysis using `pandas`.
3.  **Visualize** the long-term price evolution using `matplotlib` to identify key growth phases.

## Tech Stack
* **Python**: Core programming language.
* **yfinance**: For retrieving market data.
* **Pandas**: For data manipulation.
* **Matplotlib**: For data visualization.

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

## Data Extraction and Preprocessing

In this section, we retrieve the maximum available historical data for Nvidia from Yahoo Finance. We perform the following steps:
* Fetch the `History` of the stock with `period="max"`.
* Reset the index to make the `Date` accessible as a column.
* Ensure the `Date` column is in the correct datetime format for plotting.
* Save the raw data to a CSV file for backup/offline use.

```python
nvidia = yf.Ticker("NVDA")
nvidia_data = nvidia.history(period = "max")
nvidia_data.reset_index(inplace = True)
nvidia_data.to_csv('nvidia_data.csv', index=False)
nvidia_data["Date"] = pd.to_datetime(nvidia_data["Date"]) 
nvidia_data.head()
```

## Visualization: Price Evolution Over Time

To better understand the stock's performance, we plot the **Close Price** against the **Date**. This line chart allows us to visually inspect:
* Periods of stability or latency.
* Trends of volatility.
* The impact of recent market shifts (e.g., the AI boom).

```python
def plot_graph(df, label, color, title):
    plt.figure(figsize=(10, 5))
    plt.plot(df["Date"], df["Close"], label = label, color = color)
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel("Stock Price in USD")
    plt.legend(loc = "upper left")
    plt.grid(True)
    plt.show()
    

plot_graph(nvidia_data, "Nvidia Stock Price", "green", "Nvidia Stock Price Over Time")
```

### Observation: Nvidia Stock Price Evolution (1999–2025)

The provided line chart illustrates the historical price trajectory of Nvidia Corporation (NVDA) over a 25-year period. The visualization plots the stock price in USD on the y-axis against the timeline on the x-axis, revealing distinct phases of market behavior.
Key Analysis of Trends:

+ 1999–2015: Period of Latency:
    For the first 16 years shown, the stock price exhibits a prolonged period of stability, trading relatively flat and hovering near the \$0–$5 range (adjusted). Volatility during this phase is minimal compared to the magnitudes seen in later years, indicating a long consolidation phase before the company's major breakouts.

+ 2016–2020: The Initial Breakout:
    Around 2016, a discernible upward trend begins. The stock gradually climbs, breaking away from its historical baseline. By 2020, the price had established a new trading range, showing consistent growth driven by the gaming and early data center markets.

+ 2020–2022: Volatility and Growth:
    The period between 2020 and 2022 shows increased volatility. The stock experienced a significant rally, pushing past \$50 and approaching \$100, followed by a noticeable correction in 2022. This reflects broader market sentiments and supply chain dynamics during the post-pandemic era.

+ 2023–Present: Exponential Surge:
    The most dominant feature of the chart is the parabolic vertical ascent starting in late 2022 and continuing through 2024. The stock price skyrocketed from approximately \$50 to over \$200, representing a multi-fold increase in a very short timeframe. This "hockey stick" growth curve correlates strongly with the global boom in Generative AI, positioning Nvidia as a primary beneficiary of the demand for high-performance computing hardware.

#### Conclusion:

The graph highlights a dramatic shift in valuation dynamics. While Nvidia spent the majority of its history as a steady-growth asset, the last 24 months have been defined by hyper-growth, driven by a fundamental shift in its market relevance due to the AI revolution.
